```python
import copy

turns = {"^": ">", ">": "v", "v": "<", "<": "^"}
dx = {"^": -1, ">": 0, "v": 1, "<": 0}
dy = {"^": 0, ">": 1, "v": 0, "<": -1}

matrix = [list(row) for row in input.strip().split('\n')]
original = copy.deepcopy(matrix)

ox = 0
oy = 0
odir = ""
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        if matrix[i][j] in turns:
            ox = i
            oy = j
            odir = matrix[i][j]

ox, oy, odir
```

```python
# Answer 1
x, y, dir = ox, oy, odir
while True:
    matrix[x][y] = "X"
    nx, ny = x + dx[dir], y + dy[dir]

    if (nx) < 0 or (ny) < 0 or (nx) >= len(matrix) or (ny) >= len(matrix[0]):
        break

    if matrix[nx][ny] == "#":
        dir = turns[dir]
    else:
        x, y = nx, ny


total = 0
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        if matrix[i][j] == "X":
            total += 1
total
```

```python
# Answer 2
x, y, dir = ox, oy, odir
while True:
    matrix[x][y] = "X"
    nx, ny = x + dx[dir], y + dy[dir]

    if (nx) < 0 or (ny) < 0 or (nx) >= len(matrix) or (ny) >= len(matrix[0]):
        break

    if matrix[nx][ny] == "#":
        dir = turns[dir]
    else:
        x, y = nx, ny


def has_loop(matrix):
    x, y, dir = ox, oy, odir
    
    record = set()
    
    while True:
        if (x, y, dir) in record:
            return True

        record.add((x, y, dir))
        
        matrix[x][y] = "X"
        nx, ny = x + dx[dir], y + dy[dir]

        if (nx) < 0 or (ny) < 0 or (nx) >= len(matrix) or (ny) >= len(matrix[0]):
            break

        if matrix[nx][ny] in ["#", "O"]:
            dir = turns[dir]
        else:
            x, y = nx, ny
    return False


total = 0
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        if matrix[i][j] == "X":
            new_matrix = copy.deepcopy(original)
            new_matrix[i][j] = "O"

            if has_loop(new_matrix):
                total += 1

total
# option one: iterar todas las X, poner un objeto y detectar un loop
```

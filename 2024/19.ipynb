```python
towels = set([t.strip() for t in input.split('\n')[1].split(',')])
designs = [t.strip() for t in input.split('\n')[3:]]
```

```python
# Part 1

# Definition: dp[i] is True if design[0:i] can be formed using towels.
# Base Case: dp[0] = True (an empty design can always be formed).
# Recurrence: For each position i in the design. For each position j from 0 to i. Check if design[j:i] is in towels and dp[j] is True.
# Result: dp[len(design)] is True if the design can be formed using towels.
def is_possible(design):
    dp = [False for _ in range(len(design) + 1)]
    dp[0] = True

    for i in range(1, len(design) + 1):
        for j in range(i):
            if dp[j] and design[j:i] in towels:
                dp[i] = True
                break
    
    return dp[len(design)]

total_designs = 0
for design in designs:
    if is_possible(design):
        total_designs += 1
total_designs
```

```python
# Part 2

# Definition: dp[i] is n if design[0:i] can be formed using towels in n ways.
# Base Case: dp[0] = 1 (an empty design can always be formed in one way).
# Recurrence: For each position i in the design. For each position j from 0 to i. Check if design[j:i] is in towels and dp[j] is not 0. If so, dp[i] += dp[j].
# Result: dp[len(design)] is the number of ways the design can be formed using towels.
def count_possibilities(design):
    dp = [0 for _ in range(len(design) + 1)]
    dp[0] = 1

    for i in range(1, len(design) + 1):
        for j in range(i):
            if dp[j] and design[j:i] in towels:
                dp[i] += dp[j]
    
    return dp[len(design)]

total_designs = 0
for design in designs:
    total_designs += count_possibilities(design)
total_designs
```

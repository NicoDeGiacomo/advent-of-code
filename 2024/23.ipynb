```python
# We first parse the input and prepare the graph
graph = {}
for line in input.split('\n'):
    a, b = line.split('-')
    if a == b:
        continue
    if a not in graph:
        graph[a] = []
    if b not in graph:
        graph[b] = []
    graph[a].append(b)
    graph[b].append(a)
```

```python
# Im going to use the Bron–Kerbosch algorithm to find all maximal cliques in the graph
# https://w.wiki/CqBB

def bk(R, P, X, graph, cliques):
    if not P and not X:
        cliques.append(R)
        return

    for node in list(P):
        new_R = R | {node}  # Add the node to the current clique
        new_P = P & set(graph[node])  # Nodes that can still be part of the clique
        new_X = X & set(graph[node])  # Nodes that have already been processed
        bk(new_R, new_P, new_X, graph, cliques)  # Recurse
        P.remove(node)  # Remove the node from P
        X.add(node)  # Add it to X to exclude it from further exploration

def find_maximal_cliques(graph):
    cliques = []
    nodes = set(graph.keys())
    bk(set(), nodes, set(), graph, cliques)
    return cliques
```

```python
# Part 1
# We need to process the cliques since some that are bigger than 3 will contain smaller 3-cliques
# For example, if we have a 4-clique, it will contain 4 3-cliques.
# {'co', 'de', 'ka', 'ta'}, -> {'co', 'de', 'ka'}, {'co', 'de', 'ta'}, {'co', 'ka', 'ta'}, {'de', 'ka', 'ta'}
# We need to sort the 3-cliques to avoid duplicates inside the set. To avoid sorting we will use frozenset since it is hashable.

from itertools import combinations

cliques = find_maximal_cliques(graph)

triangular_cliques = set()
for clique in cliques:
    if len(clique) >= 3:
        for c in combinations(clique, 3):
            triangular_cliques.add(frozenset(c))

total = 0
for clique in triangular_cliques:
    if any(pc[0] == "t" for pc in clique):
        total += 1
total
```

```python
# Part 2
# Using what we have already done, we can find the largest clique, sort it, and join it with a comma.

cliques = find_maximal_cliques(graph)

largest_clique = max(cliques, key=len)

','.join(sorted(largest_clique))
```

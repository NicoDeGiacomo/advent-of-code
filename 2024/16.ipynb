```python
matrix = [list(row) for row in input.strip().split('\n')]

def find_start(matrix):
    for i, row in enumerate(matrix):
        for j, cell in enumerate(row):
            if cell == 'S':
                return i, j
```

```python
# Answer 1

directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
start = find_start(matrix)

# cell = (position, direction, score)
start_cell = (start, (0, 1), 0)

queue = [start_cell]
scores = []

memory = {}
def move(cell):
    position, direction, score = cell

    if position in memory and memory[position] <= score:
        return
    memory[position] = score

    ni, nj = position[0], position[1]
    if matrix[ni][nj] == 'E':
        scores.append(score)
        return
    if ni < 0 or ni >= len(matrix) or nj < 0 or nj >= len(matrix[0]) or matrix[ni][nj] == '#':
        return
    
    for new_direction in directions:
        if new_direction == direction:
            new_position = (position[0] + new_direction[0], position[1] + new_direction[1])
            new_score = score + 1
        else:
            new_position = (position[0] + new_direction[0], position[1] + new_direction[1])
            new_score = score + 1001
        queue.append((new_position, new_direction, new_score))


while queue:
    move(queue.pop(0))

min(scores)
```

```python
# Answer 2

directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
start = find_start(matrix)

# cell = (position, direction, score, path)
start_cell = (start, (0, 1), 0, [start])

queue = [start_cell]
scores = []

memory = {}
def move(cell):
    position, direction, score, path = cell

    if (position, direction) in memory and memory[position, direction] < score:
        return
    memory[position, direction] = score

    ni, nj = position[0], position[1]
    if matrix[ni][nj] == 'E':
        scores.append((score, path))
        return
    if ni < 0 or ni >= len(matrix) or nj < 0 or nj >= len(matrix[0]) or matrix[ni][nj] == '#':
        return
    
    for new_direction in directions:
        if new_direction == direction:
            new_position = (position[0] + new_direction[0], position[1] + new_direction[1])
            new_score = score + 1
        else:
            new_position = (position[0] + new_direction[0], position[1] + new_direction[1])
            new_score = score + 1001
        queue.append((new_position, new_direction, new_score, path + [new_position]))


while queue:
    move(queue.pop(0))

min_score = scores[0][0]
path_cells = set()
for score in scores:
    if score[0] > min_score:
        continue
    
    if score[0] < min_score:
        min_score = score[0]
        path_cells = set()
    
    if score[0] == min_score:
        for cell in score[1]:
            path_cells.add(cell)

print(min_score, len(path_cells))
```

```python
matrix = [list(row) for row in input.split('\n')]
```

```python
# Part 1

# The idea behind the algorithm is to iterate over the matrix a give each cell a unique id.
# When processing a cell, we calculate its area and perimeter. Then we pass the same id to its neighbors of the same 'type'.
# Thus identifying the connected components of the matrix, which will receive the same id.
# The 'separated' groups will have different ids, given by the original iteration. This means they will have different areas and perimeters.
#
# By storing the already visited cells we avoid unnecessary calculations, making this algorithm O(n) in time complexity (we will visit each cell only once).

sides = [(0, 1), (1, 0), (0, -1), (-1, 0)]
visited = set()
data = {}

def calculate_perimeter(i, j, id):
    if (i, j) in visited:
        return
    
    visited.add((i, j))

    if id not in data:
        data[id] = {'area': 0, 'perimeter': 0}
    
    data[id]['area'] += 1

    for side in sides:
        ni = i + side[0]
        nj = j + side[1]
        if ni < 0 or nj < 0 or ni >= len(matrix) or nj >= len(matrix[0]) or matrix[ni][nj] != matrix[i][j]:
            data[id]['perimeter'] += 1
    
    for side in sides:
        ni = i + side[0]
        nj = j + side[1]
        if ni >= 0 and nj >= 0 and ni < len(matrix) and nj < len(matrix[0]) and matrix[ni][nj] == matrix[i][j]:
            calculate_perimeter(ni, nj, id)
    
id = 0
for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        calculate_perimeter(i, j, id)
        id += 1

total = 0
for d in data:
    total += data[d]['area'] * data[d]['perimeter']
total
```

```python
# Part 2

# For this solution we need to iterate over the matrix three times.
# In the first iteration we will recognize the groups and give them a unique id, like in the previous solution.
# In the second and third iterations we will calculate the walls of the groups. We do this in two separate iterations because we need to calculate the walls in the x and y directions separately.
# To recognize the walls we use a "continuity id" that is incremented for each continuous cell. This way we can identify if the current wall was already accounted for in the previous cell.

all_sides = [(0, 1), (1, 0), (0, -1), (-1, 0)]
x_sides = [(0, 1), (0, -1)]
y_sides = [(1, 0), (-1, 0)]
sides_help = {(0, 1): 'x', (1, 0): 'y', (0, -1): 'x', (-1, 0): 'y'}

id_by_cell = {}

visited = set()
data = {}

def assign_groups(i, j, id):
    if (i, j) in visited:
        return
    
    visited.add((i, j))
    id_by_cell[(i, j)] = id

    if id not in data:
        data[id] = {'area': 0, 'perimeter': 0, 'walls': set()}
    
    data[id]['area'] += 1
    
    for side in sides:
        ni = i + side[0]
        nj = j + side[1]
        if ni >= 0 and nj >= 0 and ni < len(matrix) and nj < len(matrix[0]) and matrix[ni][nj] == matrix[i][j]:
            assign_groups(ni, nj, id)

def calculate_walls(i, j, cont_id, sides_for_walls, sides_to_walk):
    if (i, j) in visited:
        return
    
    visited.add((i, j))

    id = id_by_cell[(i, j)]
    
    for side in sides_for_walls:
        ni = i + side[0]
        nj = j + side[1]
        if ni < 0 or nj < 0 or ni >= len(matrix) or nj >= len(matrix[0]) or matrix[ni][nj] != matrix[i][j]:
            wall = (side, j if sides_help[side] == 'x' else i, cont_id - 1)
            if not wall in data[id]['walls']:
                data[id]['perimeter'] += 1
            data[id]['walls'].add((side, j if sides_help[side] == 'x' else i, cont_id))

    for side in sides_to_walk:
        ni = i + side[0]
        nj = j + side[1]
        if ni >= 0 and nj >= 0 and ni < len(matrix) and nj < len(matrix[0]) and matrix[ni][nj] == matrix[i][j]:
            calculate_walls(ni, nj, cont_id + 1, sides_for_walls, sides_to_walk)

id = 0
for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        assign_groups(i, j, id)
        id += 1

visited = set()
n_cont_id = 0
for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        calculate_walls(i, j, n_cont_id, x_sides, y_sides)
        pass
    n_cont_id += 1

visited = set()
n_cont_id = 0
for j in range(len(matrix)):
    for i in range(len(matrix[0])):
        calculate_walls(i, j, n_cont_id, y_sides, x_sides)
        pass
    n_cont_id += 1


total = 0
for d in data:
    total += data[d]['area'] * data[d]['perimeter']
total
```

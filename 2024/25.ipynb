```python
def sum_columns(s):
    return [sum([row[i] for row in s]) for i in range(len(s[0]))]

schemas = []
for s in input.split('\n\n'):
    schema = []
    for line in s.split('\n'):
        schema.append([c == '#' for c in line])
    schemas.append(schema)
schemas

locks = []
keys = []
for s in schemas:
    if s[0] == [True] * len(s[0]):
        locks.append(sum_columns(s[1:]))
        continue
    else:
        keys.append(sum_columns(s[:-1]))
```

```python
# Part 1 
# Lets try brute force
matches = 0
for key in keys:
    for lock in locks:
        if all([key[i] + lock[i] <= 5 for i in range(len(key))]):
            matches += 1
matches
```

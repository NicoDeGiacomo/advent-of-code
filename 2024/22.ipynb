```python
# This could be done with bit shifting, but I dont think it would be that much faster.
# This could also be done recursively with memoization, to save some re-calculation.
def simulate(secret, steps):
    for _ in range(steps):
        tmp = secret * 64 # calculate
        secret = secret ^ tmp # mix
        secret = secret % 16777216 # prune

        tmp = secret // 32 # calculate
        secret = secret ^ tmp # mix
        secret = secret % 16777216 # prune

        tmp = secret * 2048 # calculate
        secret = secret ^ tmp # mix
        secret = secret % 16777216 # prune

    return secret
```

```python
# Part 1
sum([simulate(buyer, 2000) for buyer in buyers])
```

```python
# We need to make some changes to the simulate function.
def simulate_seq(secret, steps):
    sequence = [int(str(secret)[-1])]
    for _ in range(steps):
        tmp = secret * 64 # calculate
        secret = secret ^ tmp # mix
        secret = secret % 16777216 # prune

        tmp = secret // 32 # calculate
        secret = secret ^ tmp # mix
        secret = secret % 16777216 # prune

        tmp = secret * 2048 # calculate
        secret = secret ^ tmp # mix
        secret = secret % 16777216 # prune

        sequence.append(int(str(secret)[-1]))

    return sequence

def calculate_differences(sequence):
    differences = []
    for i in range(1, len(sequence)):
        differences.append(sequence[i] - sequence[i-1])
    return differences
```

```python
# Part 2

bananas_per_sequence = {}
for buyer in buyers:
    prices = simulate_seq(buyer, 2000)
    diffs = calculate_differences(prices)

    # Extract sequences and calculate bananas for each sequence
    seen = set()
    for i in range(len(diffs) - 4):
        seq = tuple(diffs[i:i+4])
        if seq not in seen:
            if seq not in bananas_per_sequence:
                bananas_per_sequence[seq] = 0
            bananas_per_sequence[seq] += prices[i + 4]
            seen.add(seq)

max_bananas = 0
for seq, bananas in bananas_per_sequence.items():
    if bananas > max_bananas:
        max_bananas = bananas
max_bananas
```

```python
# Answer 2

input_list = input.split('\n')
pairs = [line.split('|') for line in input_list if '|' in line]
sequences = [line.split(',') for line in input_list if ',' in line]

# key -> number, value -> prohibited numbers after the key
rules = {}

for pair in pairs:
    if pair[1] not in rules:
        rules[pair[1]] = []
    rules[pair[1]].append(pair[0])

def is_valid_sequence(sequence):
    prohibited = set()
    for i in sequence:
        if i in prohibited:
            return False
        if i in rules:
            prohibited.update(rules[i])
    return True

def validate_and_swap(sequence):
    prohibited = set()
    for i in range(len(sequence)):
        if sequence[i] in prohibited:
            sequence[i - 1], sequence[i] = sequence[i], sequence[i - 1]
            return validate_and_swap(sequence)
        if sequence[i] in rules:
            prohibited.update(rules[sequence[i]])
    return True

total = 0
for sequence in sequences:
    if not is_valid_sequence(sequence):
        validate_and_swap(sequence)
        total += int(sequence[len(sequence) // 2])
total
```

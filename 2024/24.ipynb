```python
inputs = {}
for i in input.split('\n\n')[0].split('\n'):
    key, value = i.split(': ')
    inputs[key] = int(value)

connections = set()
for i in input.split('\n\n')[1].split('\n'):
    _in, _out = i.split(' -> ')
    _in1, _op, _in2 = _in.split(' ')
    connections.add((_in1, _in2, _op, _out))
```

```python
# Useful functions for part 2
from collections import defaultdict

connections_map = defaultdict(str)
for i in input.split('\n\n')[1].split('\n'):
    _in, _out = i.split(' -> ')
    _in1, _op, _in2 = _in.split(' ')
    connections_map[_in1, _in2, _op] = _out
    connections_map[_in2, _in1, _op] = _out

connections_input_per_input = defaultdict(str)
for i in input.split('\n\n')[1].split('\n'):
    _in, _out = i.split(' -> ')
    _in1, _op, _in2 = _in.split(' ')
    connections_input_per_input[_in2, _op] = _in1
    connections_input_per_input[_in1, _op] = _in2

connections_per_output = defaultdict(str)
for i in input.split('\n\n')[1].split('\n'):
    _in, _out = i.split(' -> ')
    _in1, _op, _in2 = _in.split(' ')
    connections_per_output[_out] = (_in1, _op, _in2)
```

```python
# Im going to code a class that helps me simulate the circuit and the logic gates.
from __future__ import annotations

class Cable:
    def __init__(self, name: str):
        self.name = name
        self.value = None
        self.connections: list[Gate] = []
    
    def connectTo(self, gate: Gate):
        self.connections.append(gate)

    def setValue(self, value: int):
        self.value = value
        for connection in self.connections:
            connection.update()

class Gate:
    def __init__(self, input1: Cable, input2: Cable, operator: str, output: Cable):
        self.input1: Cable = input1
        
        self.input2: Cable = input2
        
        self.operator = operator
        
        self.output: Cable = output

    def update(self):
        if self.input1.value is None or self.input2.value is None:
            return
        if self.operator == 'AND':
            self.output.setValue(self.input1.value & self.input2.value)
        elif self.operator == 'OR':
            self.output.setValue(self.input1.value | self.input2.value)
        elif self.operator == 'XOR':
            self.output.setValue(self.input1.value ^ self.input2.value)
```

```python
# Part 1
# Now we just have to use the classes to simulate the circuit.

# Process the connections
cables = {}
for c in connections:
    _in1, _in2, _op, _out = c
    
    # Create the cables
    if not _in1 in cables:
        cables[_in1] = Cable(_in1)
    if not _in2 in cables:
        cables[_in2] = Cable(_in2)
    if not _out in cables:
        cables[_out] = Cable(_out)

    # Create the gate
    gate = Gate(cables[_in1], cables[_in2], _op, cables[_out])

    # Connect the cables
    cables[_in1].connectTo(gate)
    cables[_in2].connectTo(gate)

# Set the inputs
for key, value in inputs.items():
    cables[key].setValue(value)

# Get the result
sorted_z_cables = sorted((cable for name, cable in cables.items() if name[0] == 'z'), key=lambda c: c.name, reverse=True)
binary_string = ''.join(str(cable.value) for cable in sorted_z_cables)
int(binary_string, base=2)
```

```python
# Part 2
# Our system is trying to perform a sum of 2 numbers, but the result is wrong.
# We have to find 4 pairs of outputs that ares swapped.
# We cant brute-force it by swapping all the outputs, we need to find the error in the circuit.

# To build and adder the circuit has to use many full-adders, and therefore should have the following structure (for each bit).
# https://media.geeksforgeeks.org/wp-content/uploads/3-57.png
#
# inputs: a, b
# carry: c
# outputs: sum, carry
#
# a XOR b -> axb
# a AND b -> aab
#
# axb XOR c -> sum
#
# axb AND c -> axbac
#
# aab OR axbac -> carry

# Lets try to find the error in the building of the circuit bite by bite.
# Lets ignore x00 and y00 since they are the carry-in of the first bit.
assert connections_map['x01', 'y01', 'XOR'] == 'bck'
assert connections_map['x01', 'y01', 'AND'] == 'cjr'

assert connections_map['bck', 'rdm', 'XOR'] == 'z01' # sum

assert connections_map['bck', 'rdm', 'AND'] == 'dsn'

assert connections_map['cjr', 'dsn', 'OR'] == 'rpn' # carry

# We can conclude that the full-adder for the first bit is correct.
```

```python
# Now lets try to implement an algorithm that finds the error for the rest of the bits.

def do_swap(a, b):
    _in1, _op, _in2 = connections_per_output[a]
    connections_map[_in1, _in2, _op] = b
    connections_map[_in2, _in1, _op] = b

    _in1, _op, _in2 = connections_per_output[b]
    connections_map[_in1, _in2, _op] = a
    connections_map[_in2, _in1, _op] = a

def check_full_adder(bit, carry):
    axb = connections_map[f'x{bit:02}', f'y{bit:02}', 'XOR']
    if not axb:
        return None, None # Not needed for this input

    aab = connections_map[f'x{bit:02}', f'y{bit:02}', 'AND']
    if not aab:
        return None, None # Not needed for this input
    
    sum_bit = connections_map[axb, carry, 'XOR']
    if not sum_bit:
        # Sum gate not found: axb AND carry. Lets find the swapped bits.
        swapped1 = connections_input_per_input[axb, 'XOR']
        swapped2 = connections_input_per_input[carry, 'XOR']

        # The input that find a gate is the one that is correct.
        errors = [swapped1, carry] if swapped1 else [swapped2, axb]
        return None, errors

    axbac = connections_map[axb, carry, 'AND']
    if not axbac:
        return None, None # Not needed for this input
    
    carry_bit = connections_map[aab, axbac, 'OR']
    if not carry_bit:
        # Carry gate not found: aab OR axbac. Lets find the swapped bits.
        swapped1 = connections_input_per_input[aab, 'OR']
        swapped2 = connections_input_per_input[axbac, 'OR']

        # The input that find a gate is the one that is correct.
        errors = [swapped1, axbac] if swapped1 else [swapped2, aab]
        return None, errors

    return carry_bit, None
```

```python
# Part 2

bit = 1 # We ignore the first bit because the structure is different
prev_carry = connections_map['y00', 'x00', 'AND'] # Carry for bit 1
errors = []

# We process each full-adder for each bit. If we find an error we swap the bits and start over.
while bit < 44:
    prev_carry, error = check_full_adder(bit, prev_carry)
    if error:
        errors.append(error)
        do_swap(error[0], error[1])

        bit = 1
        prev_carry = connections_map['y00', 'x00', 'AND']
        continue

    bit += 1

# Flatten, sort, and concatenate the errors
','.join(sorted(e for inner in errors for e in inner))
```
